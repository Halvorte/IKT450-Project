```python
# Import
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
#from pandas.table.plotting import table
```

```python
# Import the data
stock_df = pd.read_csv('stock_data/AAPL.csv')
stock_amzn = pd.read_csv('stock_data/AMZN.csv')
stock_orcl = pd.read_csv('stock_data/ORCL.csv')
```

```python
# Display orcl stock info
stock_orcl.info()
display(stock_orcl)
```

```python
# Plot the data
# Plot closing value
stock_plt = stock_df
stock_plt['Date'] = pd.to_datetime(stock_plt['Date'])
display(stock_plt['Date'])
plt.figure(figsize=(22,10))
plt.plot(stock_plt.Date, stock_plt.Close)
plt.title("Closing value of stock")
plt.xlabel("Date")
plt.ylabel("Closing value")
plt.show()
```

```python
# Plot the volume for each day
stock_vol = stock_df
stock_vol['Date'] = pd.to_datetime(stock_vol['Date'])
plt.figure(figsize=(22,10))
plt.plot(stock_vol.Date, stock_vol.Volume)
plt.title("Volume of stock")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
fig, axs = plt.subplots(2)
fig.suptitle('AAPL: Closing value and Volume')
axs[0].plot(stock_plt.Date, stock_plt.Close)
axs[1].plot(stock_vol.Date, stock_plt.Volume, color='Maroon')
fig.savefig('aapl_close_vol.png')
```
